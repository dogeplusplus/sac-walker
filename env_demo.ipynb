```python
import brax
import jax
import jax.numpy as jnp

from brax import envs
from brax.io import html
from IPython.display import HTML
```

```python
def visualize(sys, qps):
    return HTML(html.render(sys, qps))

def main():
    env_name = "humanoid"
    env_fn = envs.create_fn(env_name)
    env = env_fn()
    jit_env_reset = jax.jit(env.reset)
    state = jit_env_reset(rng=jax.random.PRNGKey(0))
    visualize(env.sys, [state.qp])
```

```python
def inference_fn(observations, seed):
    return jax.random.uniform(seed, shape=(17,))
```

```python
env_name = "humanoid"
env_fn = envs.create_fn(env_name)
env = env_fn()
jit_env_reset = jax.jit(env.reset)
```

```python
jit_env_step = jax.jit(env.step)
jit_inference_fn = jax.jit(inference_fn)
state = jit_env_reset(rng=jax.random.PRNGKey(0))
qps = []
while not state.done:
    qps.append(state.qp)
    act = jit_inference_fn(state.obs, state.rng)
    state = jit_env_step(state, act)
    
visualize(env.sys, qps)
```
